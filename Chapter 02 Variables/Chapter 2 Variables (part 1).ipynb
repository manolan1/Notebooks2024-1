## What Is a Variable?

### `a = "Hello World"`



```python
a = "Hello World"
```

```python
a
```

```python
type(a)
```

The `type()` function returns the type (or _class_) of its argument, in this case `str`.

```python
id(a)
```

The built-in function `id()` returns the _identity_ of an object
- This is an integer that is guaranteed to be constant and unique over the lifetime of the object. 
- Two objects with non-overlapping lifetime may have the same id.
- In many Python implementations, this is the address in memory.

### Objects, Identifiers, References, and Variables

- Everything is an object
  - The symbol table keeps track of which variables are in use
    - The variable name, like `a`, is known as an _identifier_ 
  - Each object counts the number of variables pointing to it
- Python identifiers consist of a combination of upper and lower case letters, digits and the underscore character
  - Cannot start with a digit
  - Combinations of leading and trailing underscores have specific meaning
    - Some by definition, some by convention
- When a new value is assigned, the reference count of the old object is decremented
  - When the reference count is 0, objects will be cleared up by garbage collection


```python
a = 4242
```

```python
type(a)
```

```python
id(a)
```

### Objects, Identifiers, References, and Variables (continued)

- More than one variable can refer to the same object
  - Reference count keeps track of how many
- The operator `is` can be used to identify this situation



```python
b = a
```

```python
b
```
